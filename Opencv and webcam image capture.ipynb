```python
import cv2
import matplotlib.pyplot as plt
```

```python
def main():
    cap=cv2.VideoCapture(0)
     
    if cap.isOpened():
        ret,img=cap.read()
        print(img)
        print(ret)
    else:
        ret=False
        
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
     
    plt.imshow(img1)
    plt.title('Color Image BGR')
    plt.xticks([])
    plt.yticks([]) 
    plt.show()
    
    
    cap.release()
    
    
if __name__=='__name__':
    main()
```

```python
main()
```
