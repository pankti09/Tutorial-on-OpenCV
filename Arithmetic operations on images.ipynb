```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def main():
    
    path='C:\\Users\\SHAILESH SHAH\\Documents\\opencv\\misc\\'
    imgpath1 = path + '4.1.03.tiff'
    imgpath2 = path + '4.1.01.tiff'
    
    img1=cv2.imread(imgpath1,1)
    img2=cv2.imread(imgpath2,1)
    
    img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
    img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)        
    
    add=img1+img2
    
    add1=img1 + 50
    
    sub1=img1-img2
    
    titles=['Roma','Arina','Addition','Addition1','img1-img2']
    images=[img1,img2,add,add1,sub1]
    
    for i in range(5):
        plt.subplot(1,5,i+1)
        plt.imshow(images[i])
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
    plt.show()
```

```python
main()
```
