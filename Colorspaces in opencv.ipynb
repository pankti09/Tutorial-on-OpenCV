```python
import cv2
import matplotlib.pyplot as plt

def main():
    
    imgpath="C:\\Users\\SHAILESH SHAH\\Documents\\opencv\\misc\\4.1.01.tiff"
    img=cv2.imread(imgpath,0)
    
    plt.imshow(img,cmap='gray')
    plt.title('Gray colormap')
    plt.show()
    
    plt.imshow(img)
    plt.title('Default colormap')
    plt.show()
    
    plt.imshow(img,cmap='Spectral_r')
    plt.title('Spectral_r colormap')
    plt.show()
    
    plt.imshow(img,cmap='Spectral')
    plt.title('Spectral colormap')
    plt.show()
    
    plt.imshow(img,cmap='Blues')
    plt.title('Blues colormap')
    plt.show()
    
    plt.imshow(img,cmap='Reds')
    plt.title('Reds colormap')
    plt.show()
    
    plt.imshow(img,cmap='Accent')
    plt.title('Accent colormap')
    plt.show()
    
    
    cv2.imshow('Arina',img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
if __name__=="__main__" :
    main()
```

```python
import cv2
import matplotlib.pyplot as plt

def main():
    
    imgpath="C:\\Users\\SHAILESH SHAH\\Documents\\opencv\\misc\\4.1.01.tiff"
    img=cv2.imread(imgpath,1)
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    
    plt.imshow(img)
    plt.title('Default  colormap')
    plt.xticks([])
    plt.yticks([])
    plt.show()
    
    plt.imshow(img1)
    plt.title('BGR 2 RGB colormap')
    plt.xticks([])
    plt.yticks([])
    plt.show()
    
    
    
if __name__=="__main__":
    main()
```
